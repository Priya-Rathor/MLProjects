```python
import numpy as np
import pandas as pd 
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')

ipl = delivery.merge(match,left_on='match_id',right_on= 'id')
ipl.head()
```

# 1. Scatter Plots

 1.Scatter plots are drown bwtween to continous variables

 Problem :- We are going to draw a scatter plot between batsman avg(x axis) and 

 Batsman strike Rate (Y axis) of thje top 50 batsman in IPL (all time)



```python
print(ipl.columns)
```

```python
top50 = ipl.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(50).index.tolist()
new_ipl = ipl[ipl['batter'].isin(top50)]
```

```python
# Calculating SR
# SR = [(number of runs scored)/(number of balls played)] * 100

runs = new_ipl.groupby('batter')['batsman_runs'].sum()
balls = new_ipl.groupby('batter')['batsman_runs'].count()

sr = (runs/balls)*100

sr =sr.reset_index()
sr
```

```python
# Calculating Avg
# Avg = (Total number of Runs)/(number of outs)
# calculating number of outs of for tiop 50 batsman

out = ipl[ipl['player_dismissed'].isin(top50)]
nouts = out['player_dismissed'].value_counts()

avg = runs/nouts

avg = avg.reset_index()

avg.rename(columns={'index':'batter',0:'avg'}, inplace=True)

avg = avg.merge(sr,on='batter')
```

```python
# Plot Scatter Plot here
trace = go.Scatter(x=avg['avg'],y=avg['batsman_runs'],mode='markers',text=avg['batter'] , marker={'color':"#00a65a",'size':16})
data = [trace]

layout = go.Layout(title='batsman Avg vs SR', xaxis = {'title':'Batsman Average'},yaxis={'title':'Batsman Strike rate' })
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

# 2.Line Chart

It's an extension of Scatter plot. Usually . used to  show a time series data

```python
# Year by year batsman performance

single = ipl[ipl['batter'] == 'V Kohli']
performance = single.groupby('season')['batsman_runs'].sum().reset_index()
performance

single1 = ipl[ipl['batter'] == 'MS Dhoni']
performance1 = single1.groupby('season')['batsman_runs'].sum().reset_index()
performance1
```

```python
# Plot Line chart here 
trace = go.Scatter(
    x=performance['season'],
    y=performance['batsman_runs'],
    mode='lines',  # Corrected from 'node' to 'mode'
    marker=dict(color='#00a65a')  # Corrected marker syntax
)


trace1 = go.Scatter(
    x=performance1['season'],
    y=performance1['batsman_runs'],
    mode='lines',  # Corrected from 'node' to 'mode'
)

data = [trace,trace1]

layout = go.Layout(title='Year by Year Performance',xaxis={'title':'Season'},yaxis={'title':'Total runs'})

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig, filename='myfile.html')
```

```python
# Multiple Line Charts

def batsman_comp(*name):
    data = []
    for i in name:
        single = ipl[ipl['batter']==i]
        performance=single.groupby('season')['batsman_runs'].sum().reset_index()

        trace = go.Scatter(x=performance['season'],y=performance['batsman_runs'],mode='lines + markers',name=i)

        data.append(trace)

    layout = go.Layout(title='Batsman Record comparator',xaxis={'title':'Season'},yaxis={'title':'Runs'})

    fig = go.Figure(data=data, layout=layout)

    pyo.plot(fig, filename='Year_by_year.html')
```

```python
batsman_comp('V Kohli','RG Sharma','DA Warner','MS Dhni')
```

# 3 Bar Plot

```python
top10 = ipl.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10).index.tolist()
top10_df = ipl[ipl['batter'].isin(top10)]
```

```python
top10_score = top10_df.groupby('batter')['batsman_runs'].sum().reset_index()
top10_score
```

```python
# plot bar Graph

trace=go.Bar(x=top10_score['batter'], y=top10_score['batsman_runs'])

data = [trace]
layout=go.Layout(title='Top 10 IPL batsman',
                 xaxis={'title':'Batter'},
                 yaxis={'title':'Total Runs'})

fig= go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

# There are two type of Bar Graphs
1. Nested Bar Graph

2.stacked Bar Graph

3. Overlayed Bar Graph

```python
iw = top10_df.groupby(['batter','inning'])['batsman_runs'].sum().reset_index()
mask = iw['inning'] == 1
mask2=iw['inning'] == 2
one = iw[mask]
two = iw[mask2]

one.rename(columns={'batsman_runs':'1st Innings'},inplace=True)
two.rename(columns={'batsman_runs':'2nd Innings'},inplace =True)

final = one.merge(two, on='batter')[['batter','1st Innings','2nd Innings']]

final
```

```python
trace1 = go.Bar(x= final['batter'],y=final['1st Innings'],name='1st innings',marker={'color':'#00a65a'})

trace2 = go.Bar(x=final['batter'],y=final['2nd Innings'],name='2nd inninga',marker={'color':'#a6a65a'})

data = [trace1,trace2]

layout = go.Layout(title='Inning wise scores',
                   xaxis={'title':'batter'},
                   yaxis={'title':'Runs'},
                   barmode='overlay')

# barmode = 'stack'
# barmode = not any things

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

# Box Plot 

A box and whisker plot -- also called a box plot --displays the five -number summary of a set of data.

```python
match_agg = delivery.groupby(['match_id'])['total_runs'].sum().reset_index()
season_wise = match_agg.merge(match, left_on='match_id',right_on='id')[['match_id','total_runs','season']]
season_wise
```

```python
trace= go.Box(x=season_wise['total_runs'],name='All seasons')

data=[trace]

layout=go.Layout(title='Total core analysis',
                 xaxis={'title':'Total Score'})

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig)
```

```python
trace1 = go.Box(x=season_wise[season_wise['season']==2017]['total_runs'],name="2017",marker={'color':'#00a65a'})
trace2 = go.Box(x=season_wise[season_wise['season']==2008]['total_runs'],name="2008")

data=[trace1,trace2]

layout=go.Layout(title='Total score analysis',
                 xaxis={'title':'Total Score'})

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig)
```

# Histograms

A histogram is a plot that lets you discover, and show the underlying frequency distribution(shape) of a set continuous data

```python
x = delivery.groupby('batter')['batsman_runs'].count()>150
x=x[x].index.tolist()

new =delivery[delivery['batter'].isin(x)]

runs = new.groupby('batter')['batsman_runs'].sum()

balls = new.groupby('batter')['batsman_runs'].count()

sr = (runs/balls)*100
sr = sr.reset_index()

sr
```

```python
trace = go.Histogram(x=sr['batsman_runs'])

data = [trace]

layout = go.Layout(title='Strike rate analysis',
                   xaxis={'title':'strike rates'})

fig= go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

# Heatmaps

A heat map is agraphical representation of data where the individual values contained in a matrix are represented as colors

```python
six = delivery[delivery['batsman_runs']==6]

six = six.groupby(['batting_team','over'])['batsman_runs'].count().reset_index()

six
```

```python
trace= go.Heatmap(x=six['batting_team'],y=six['over'],z=six['batsman_runs'])

data = [trace]
layout= go.Layout(title='six Heatmap')

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig)
```

```python
dots = delivery[delivery['batsman_runs']==0]
dots= dots.groupby(['batting_team','over'])['batsman_runs'].count().reset_index()

from plotly import tools
trace1 = go.Heatmap(x=six['batting_team'], y=six['over'],z=six['batsman_runs'].values.tolist())

trace2 = go.Heatmap(x=dots['batting_team'],y=dots['over'],
                    z=dots['batsman_runs'].values.tolist())

fig = tools.make_subplots(rows = 1,cols=2,subplot_titles=["6's","0's"],shared_yaxes=True)

fig.append_trace(trace1,1,1)
fig.append_trace(trace2,1,2)

pyo.plot(fig)
```
